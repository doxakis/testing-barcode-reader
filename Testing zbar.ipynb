# Testing zbar

- First, you need to install zbar
- On Mac OS X: ```brew install zbar```

````python
import subprocess
from IPython.display import Image
````

````python
# No barcode here
output = subprocess.getoutput("zbarimg files/form1.jpg")
print(output)
Image(filename='files/form1.jpg')
````

````python
# Barcode here, but too small
output = subprocess.getoutput("zbarimg files/form2.jpg")
print(output)
Image(filename='files/form2.jpg')
````

````python
# 2 barcode here
output = subprocess.getoutput("zbarimg files/multiple_barcode.png")
print(output)

# 2 barcode found!

Image(filename='files/multiple_barcode.png')
````

````python
# Many barcode type here
output = subprocess.getoutput("zbarimg files/many-barcode-type.png")
print(output)

# Disappointing.. only 2 barcode found

Image(filename='files/many-barcode-type.png')
````
